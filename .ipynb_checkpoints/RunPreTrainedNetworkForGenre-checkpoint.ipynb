```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.optim import lr_scheduler
import numpy as np
import torchvision
from torchvision import datasets, models, transforms
import matplotlib.pyplot as plt
import pandas as pd
import pickle
from PIL import Image,ImageOps
import PIL
from sklearn.metrics import confusion_matrix, f1_score, plot_confusion_matrix, precision_score
import time
import os
import copy 

plt.ion()   # interactive mode
```

```python
# Create a dataset class that will handle the images we're going to predict on
class Dataset(torch.utils.data.Dataset):
  'Characterizes a dataset for PyTorch'
  def __init__(self, input_file, transforms_dict=None, train=False):
        self.data = np.array([('Images/'
                      +str(index)+'.jpg', row['genre_code']) for index,row in input_file.iterrows()])
        self.transforms_dict = transforms_dict
        # Create a list to store the paths of the images that are in the dataset.
        self.used_images = list()
        
        for tup in self.data:
            assert os.path.isfile(tup[0])
                
  def __len__(self):
        'Denotes the total number of samples'
        return len(self.data)

  def __getitem__(self, index):
        'Generates one sample of data'
        # Select sample
        tup = self.data[index]
        ID = tup[0]
        
        image = Image.open(open(ID, 'rb'))
        image = self.transforms_dict['regular'](image)
        
        self.used_images.append(ID)

        # Load data and get label
        label = int(tup[1])

        return image, label
    
  def remove(self, remove_list):
        mask = np.ones(len(self.data), dtype=bool)
        mask[remove_list] = False
        self.data = self.data[mask]
    
def get_image(image_path, reshape_size = (224,224)):
    transform = transforms.Compose([transforms.Resize(reshape_size),
    transforms.ToTensor(), transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])])
    
    image = Image.open(open(image_path, 'rb'))
    if image.mode == 'RGB':
        try:
            return transform(image)
        except:
            return None
    else:
        return None
    
def imshow(inp, title=None):
    """Imshow for Tensor."""
    inp = inp.numpy().transpose((1, 2, 0))
    mean = np.array([0.485, 0.456, 0.406])
    std = np.array([0.229, 0.224, 0.225])
    inp = std * inp + mean
    inp = np.clip(inp, 0, 1)
    plt.imshow(inp)
    if title is not None:
        plt.title(title)
    plt.pause(0.001)  # pause a bit so that plots are updated
```

```python
# Create a dictionary of transformations.
# This one only contains a transformation for normalisation but others could be added e.g. shearing, rotation etc.
transforms_dict = {'regular': transforms.Compose([transforms.Resize(224),
    transforms.ToTensor(),          
transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])   
}
```

```python
# Load the dataset created by Oramas et al.
oramas_data = pd.read_csv('oramas_data_with_names.csv')
# Convert the 'genre' field to a category
oramas_data['genre'] = oramas_data["genre"].astype('category')
# If we pass the label as text to the train set it causes an error when we try to iterate on it, so create a numeric code
oramas_data['genre_code'] = oramas_data['genre'].cat.codes

print(oramas_data[['genre', 'genre_code']])
# Create an array that we can index using the 'genre_code' variable to get the genre names
classes = oramas_data['genre'].cat.categories
```

```python
oramas_data.columns.values
```

```python
# Get the training and test sets from the Oramas et al. data and use these to create Dataset objects
training = oramas_data[oramas_data['set'] == 'train'].copy()
print(training.shape)
testing = oramas_data[oramas_data['set'] == 'test'].copy()
print(testing.shape)

train_set = Dataset(training, transforms_dict = transforms_dict, train=True)

test_set = Dataset(testing, transforms_dict = transforms_dict)
```

```python
# Most of the images are of the size (3,224,224) but a few are not. These will cause the model to break so we
# remove them. There are also some images that threw an error when shape was called on them, so we remove these too
bad_train_images = list()
good_size = torch.Size([3, 224, 224])
for i in range(len(train_set)):
    try:
        if train_set[i][0].shape != good_size:
            bad_train_images.append(i)
    except:
        bad_train_images.append(i)
    # Maintain a counter so we know how many images we've processed    
    if i % 1000 == 0:
        print(i)
```

```python
# We do the same for the test images as we did for the
bad_test_images = list()
good_size = torch.Size([3, 224, 224])
for i in range(len(test_set)):
    try:
        if test_set[i][0].shape != good_size:
            bad_test_images.append(i)
    except:
        bad_test_images.append(i)
    if i % 1000 == 0:
        print(i)
```

```python
# Remove the problem images from the train and test sets
train_set.remove(bad_train_images)
test_set.remove(bad_test_images)
```

```python
# Create data loaders for the train and test sets
trainloader = torch.utils.data.DataLoader(train_set, batch_size=64,
                                          shuffle=True, num_workers=0)

testloader = torch.utils.data.DataLoader(test_set, batch_size=64,
                                          shuffle=True, num_workers=0)
```

```python
# The used_images array is used to map the images onto their file path. This mapping only works if the array is empty before
# we iterate. We iterated over the train and test set to remove bad images. This filled up the used_images arrays in 
# sequential order. We're going to iterate over them randomly for training and testing so we need to empty the lists 
# again
test_set.used_images = list()
train_set.used_images = list()
```

```python
# Load a pretrained resnet model
# There are a lot of other models to try, but they all produce roughly the same performance
model_name = 'resnet'

device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")

model_ft = models.resnet18(pretrained=True)
num_ftrs = model_ft.fc.in_features

model_ft.fc = nn.Linear(num_ftrs, len(classes))

model_ft = model_ft.to(device)
print(model_ft.eval())

criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(model_ft.parameters(), lr=0.001, momentum=0.9)
```

```python
# Now we predict on the test set and compute accuracy
correct = 0
total = 0

all_ground_truth = list()
all_predicted = list()

with torch.no_grad():
    for data in testloader:
        images, labels = data
        outputs = model_ft(images)
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        print(total)
        correct += (predicted == labels).sum().item()
        all_ground_truth += list(labels)
        all_predicted += list(predicted)

print('Accuracy of the network on the test images: %d %%' % (
    100 * correct / total))
```

```python
# Create ground truth and predicted labels by mapping the codes back to the label names
ground_truth_labels = [classes[e.tolist()] for e in all_ground_truth]
all_predicted_labels = [classes[e.tolist()] for e in all_predicted]
```
