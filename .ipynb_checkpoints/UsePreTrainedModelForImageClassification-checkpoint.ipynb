```python
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
import pickle
from PIL import Image
import random
import sys
import torch
import torch.nn as nn
from torchvision import models, transforms

plt.ion()   # interactive mode
```

```python
transforms_dict = {'regular': transforms.Compose([transforms.Resize(size),
    transforms.ToTensor(),          
transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])   
}
```

```python
class Dataset(torch.utils.data.Dataset):
  'Characterizes a dataset for PyTorch'
  def __init__(self, input_file, transforms_dict=None, train=False):
        self.data = np.array([('C:/Users/amcgo/Documents/Images_To_Music/Music_Image_Matcher/Images/'
                      +str(index)+'.jpg', row['genre_code']) for index,row in input_file.iterrows()])
        self.transforms_dict = transforms_dict
        
        for tup in self.data:
            assert os.path.isfile(tup[0])
                
        
  def __len__(self):
        'Denotes the total number of samples'
        return len(self.data)

  def __getitem__(self, index):
        'Generates one sample of data'
        # Select sample
        tup = self.data[index]
        ID = tup[0]
        
        if ID.endswith('_rotated'):
            ID = ID.replace('_rotated','')
            image = Image.open(open(ID, 'rb'))
            image = self.transforms_dict['rotate'](image)
            
        elif ID.endswith('_mirrored'):
            ID = ID.replace('_mirrored','')
            image = Image.open(open(ID, 'rb'))
            image = self.transforms_dict['mirror'](image)
        
        elif ID.endswith('_cropped'):
            ID = ID.replace('_cropped','')
            image = Image.open(open(ID, 'rb'))
            image = self.transforms_dict['crop'](image)
        
        elif ID.endswith('_sheared'):
            ID = ID.replace('_sheared','')
            image = Image.open(open(ID, 'rb'))
            image = self.transforms_dict['shear'](image)
            
        elif ID.endswith('_color_jittered'):
            ID = ID.replace('_color_jittered','')
            image = Image.open(open(ID, 'rb'))
            image = self.transforms_dict['color_jitter'](image)
        else:
            image = Image.open(open(ID, 'rb'))
            image = self.transforms_dict['regular'](image)

        # Load data and get label
        label = int(tup[1])

        return image, label
    
  def remove(self, remove_list):
        mask = np.ones(len(self.data), dtype=bool)
        mask[remove_list] = False
        self.data = self.data[mask]
        
  def get(self, index):
      tup = self.data[index]
      ID = tup[0]
      image = Image.open(open(ID, 'rb'))
      image = self.transforms_dict['regular'](image)
      label = int(tup[1])
      return image, label
    
def imshow(inp, title=None):
    """Imshow for Tensor."""
    inp = inp.numpy().transpose((1, 2, 0))
    mean = np.array([0.485, 0.456, 0.406])
    std = np.array([0.229, 0.224, 0.225])
    inp = std * inp + mean
    inp = np.clip(inp, 0, 1)
    plt.imshow(inp)
    if title is not None:
        plt.title(title)
    plt.pause(0.001)  # pause a bit so that plots are updated
```

```python
oramas_data = pd.read_csv('oramas_data_with_names.csv')
oramas_data['genre'] = oramas_data["genre"].astype('category')
oramas_data['genre_code'] = oramas_data['genre'].cat.codes

print(oramas_data[['genre', 'genre_code']])

classes = oramas_data['genre'].cat.categories
```

```python
training = oramas_data[oramas_data['set'] == 'train'].copy()
print(training.shape)
testing = oramas_data[oramas_data['set'] == 'test'].copy()
print(testing.shape)

train_set = Dataset(training, transforms_dict = transforms_dict, train=True)

test_set = Dataset(testing, transforms_dict = transforms_dict)
```
