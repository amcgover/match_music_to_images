```python
import h5py
import pandas as pd
import pickle
```

```python
analysis_list = h5py.File('msd_summary_file.h5', 'r')['analysis']['songs']['track_id']
song_list = h5py.File('msd_summary_file.h5', 'r')['metadata']['songs']['title', 'artist_name', 'release']
```

```python
records = list()
for i in range(len(analysis_list)):
    
    records.append([analysis_list[i].decode(), song_list[i][0].decode(), 
                    song_list[i][1].decode(), song_list[i][2].decode()])
    
    if i % 10000 == 0:
        print(i)
        
song_frame = pd.DataFrame(records, columns = ['msd_track_id', 'song_name', 'artist_name', 'album_name'])
```

```python
oramas_dataset = pd.read_csv('Oramas_Data/MSD-I_dataset.tsv', delimiter = '\t')
```

```python
merged = oramas_dataset.merge(song_frame, on = 'msd_track_id')
```
