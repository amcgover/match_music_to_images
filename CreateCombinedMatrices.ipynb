```python
import numpy as np
import os
import pandas as pd
import pickle
import torch
```

```python
# Declare the convolution layers for each of the style matrices
conv_layers = ['conv_1', 'conv_2', 'conv_3', 'conv_4', 'conv_5']
```

```python
# Read in the Oramas data
oramas_data = pd.read_csv('oramas_data_with_names.csv')
```

```python
oramas_data[oramas_data['album_index'] == 0]
```

```python
# Create a loop that will iterate through each image and combine its five style matrices into a single matrix
count = 0
for album_index in oramas_data['album_index'].unique():
    # If we've already created a combined style matrix for this album index, skip
    if os.path.exists('combined_style_matrices_15_pca/%s' % album_index):
        count += 1
        continue
    # If we don't have a content matrix for the image then there was an error and we don't want to create
    # a combined style matrix
    if not os.path.exists('content_matrices_50_pca/%s_conv_4' % str(album_index)):
        continue
    # Create a list to store each of the style matrices
    album_matrices = list()
    # Create a loop to add each of the style matrices into the album_matrices list
    for c in conv_layers:
        album_matrices.append(pickle.load(open('style_matrices_15_pca/' + str(album_index) + '_' + c, 'rb')))
    # Concatenate the five style matrices into one single matrix and save
    torch.save(np.concatenate(album_matrices), 'combined_style_matrices_15_pca/%s' % album_index)
    
    # Maintain a count to track progress
    count += 1
    if count % 100 == 0:
        print(count)
```

```python
# Put the album indexes in a list. We will save this list along with the content and style matrices
album_indexes = oramas_data['album_index'].unique()
```

```python
# For every image, load the content matrices and combined style matrices, unroll them so they 
# are a single row and and save them
# Create lists for:
    # the unrolled content matrices
    # the unrolled style matrices
    # the album indexes associated with the unrolled content and style matrices
total_content_matrix,total_style_matrix,used_album_indexes = list(),list(),list()
count = 0
for index in album_indexes:
    # If we don't have a content matrix for this album index there was an error, so we skip
    if not os.path.exists('content_matrices_50_pca/%s_conv_4' % index):
        continue
    # Append the album index    
    used_album_indexes.append(index)
    # Load the content matrix, unroll it to be a single row and save it
    content_matrix = pickle.load(open('content_matrices_50_pca/%s_conv_4' % index, 'rb'))
    total_content_matrix.append(np.reshape(content_matrix, (6400,)))
    # Load the style matrix, unroll it to be a single row and save it
    style_matrix = torch.load('combined_style_matrices_15_pca/%s' % index)
    total_style_matrix.append(np.reshape(style_matrix, (9600,)))
    # Maintain a count to track progress
    count += 1
    if count % 100 == 0:
        print(count)
# Save the album indexes, total content matrices and total style matrices
pickle.dump(total_content_matrix, open('all_content_matrices_50_pca', 'wb'))
pickle.dump(total_style_matrix, open('all_style_matrices_15_pca', 'wb'))
pickle.dump(used_album_indexes, open('all_album_indexes_new_pca', 'wb'))
```
