```python
import fileinput
import itertools
import matplotlib.pyplot as plt
import numpy as np
import operator
from PIL import Image
import re
import torchvision.transforms as transforms
```

```python
def imshow(inp, title=None):
    """Imshow for Tensor."""
    inp = inp.numpy().transpose((1, 2, 0))
    mean = np.array([0.485, 0.456, 0.406])
    std = np.array([0.229, 0.224, 0.225])
    inp = std * inp + mean
    inp = np.clip(inp, 0, 1)
    plt.imshow(inp)
    if title is not None:
        plt.title(title)
    plt.pause(0.001)  # pause a bit so that plots are updated
    
def get_image(image_path):
    transform = transforms.Compose([transforms.Resize((224, 224)),
    transforms.ToTensor(), transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])])
    
    image = Image.open(open(image_path, 'rb'))
    if image.mode == 'RGB':
        try:
            return transform(image)
        except:
            return None
    else:
        return None
```

```python
line_pattern = '(\d+)\t(\d\.\d+)'

ethnicities = ['Black', 'East Asian', 'Indian', 'Latino_Hispanic', 'Middle Eastern', 'Southeast Asian',
                   'White']
# Create a dictionary to store the ethnicities and their top 80% of matches
ethnicity_to_images = dict()

for ethnicity in ethnicities:
    fileinput.close()
    # Create an entry in the dictionary for the ethnicity with an empty list
    ethnicity_to_images[ethnicity] = list()
    # Create a variable to store the cumulative match percentage
    cumulative_score = 0
    # Open the image count file and iterate through it
    for line in fileinput.input('%s_image_counts.txt' % ethnicity):
        line = line.rstrip()
        # Split the line on '\t'. Keep the image index and the percentage match
        image_number,_,score = line.split('\t')
        # Add the percentage match to the cumulative score
        cumulative_score += float(score)
        # If the cumulative score is above 80% break out of the loop
        if cumulative_score > 0.8:
            break
        # If the cumulative score is below 80% add the image index to the list for the ethnicity
        ethnicity_to_images[ethnicity].append(image_number)
```

```python
# Get the top images for each ethnicity and get the intersection
image_sets = [set(ethnicity_to_images[ethnicity]) for ethnicity in ethnicities]
# Get the intersection of the matches for every ethnicity
shared_images = set.intersection(*image_sets)

# Combine the matches for every ethnicity into one list
image_lists = ethnicity_to_images.values()
image_lists = itertools.chain(*image_lists)

# Get the images that are not in the intersection.
images_not_in_intersection = set(image_lists) - shared_images


# Create a dictionary to store the matched images and the number of ethnicities they have been matched with
images_to_ethnicity_counts = dict()
# Iterate through the images not in the intersection and track the ethnicities they have been matched with
for image in images_not_in_intersection:
    images_to_ethnicity_counts[image] = list()
    for ethnicity,ethnicity_images in ethnicity_to_images.items():
        if image in ethnicity_images:
            images_to_ethnicity_counts[image].append(ethnicity)
            
```

```python
# Of the images that aren't in the intersection, get the percentage that were matched withPr 2 or more ethnicities
# Also create a dictionary of the ethnicity combinations that images have matched with and the number of times
# they have been matched
ethnicity_to_counts = dict()
num_at_or_above = 0
for image_counts in images_to_ethnicity_counts.values():
    if len(image_counts) >= 2:
        num_at_or_above += 1
        image_counts.sort()
        image_counts = tuple(image_counts)
        if image_counts in ethnicity_to_counts:
            ethnicity_to_counts[image_counts] += 1
        else:
            ethnicity_to_counts[image_counts] = 1
        
print('Proportion of images that are matched to 2 or more ethnicities: %f' % (num_at_or_above/len(images_to_ethnicity_counts)))
```

```python
# Print the ethnicity combinations and number of images matched to them in sorted order, highest first
for e in sorted(ethnicity_to_counts.items(), key=operator.itemgetter(1), reverse=True):
    print(e)
```

```python
# Get the images that have only been matched to one ethnicity
only_with_one_ethnicity = [(image,ethnicities) for image,ethnicities in images_to_ethnicity_counts.items() if len(ethnicities) == 1]
```

```python
# Print each of the images that have only been matched to one ethnicity and the ethnicity they've been matched to
only_with_one_ethnicity
```

```python
# Display each of the images and the ethnicity they matched with
for image_name,ethnicity in only_with_one_ethnicity:
    matched_image = get_image('Images/%s.jpg' % image_name)
    imshow(matched_image)
    print('Matched with %s' % ethnicity)
```
